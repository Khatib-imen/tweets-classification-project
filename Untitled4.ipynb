
# etape 1 Prétraitement des Tweets
* Dans cette étape, l’objectif est d’éliminer le texte inutile des tweets tels que les #, les noms des utilisateurs, les URLs, etc. Cela va nous permettre de nettoyer les données avant de les analyser..*

Check the Column Names:*texte en italique*

````python
print(df.columns)
````

````python
import re

# Function to clean the tweet text
def clean_text(text):
    # Remove URLs
    text = re.sub(r'http\S+', '', text)
    # Remove mentions (usernames starting with @)
    text = re.sub(r'@\S+', '', text)
    # Remove hashtags
    text = re.sub(r'#\S+', '', text)
    # Remove any non-alphabetic characters
    text = re.sub(r'[^A-Za-z\s]', '', text)
    text = text.lower()
    return text

# Apply the cleaning function to the 'Text' column
df['cleaned_text'] = df['Text'].apply(clean_text)

# Display the first few rows to check the results
df.head()
````

#tape 2 Traitement des Tweets : NLP (Natural Language Processing)
*Cette étape consiste à analyser les tweets en utilisant la bibliothèque NLTK. Nous allons procéder à la tokenisation, qui est le processus de découper le texte en mots individuels.*

Download the punkt *resource*

````python
import nltk

# Download the punkt tokenizer data
nltk.download('punkt')

# Download punkt_tab for sentence tokenization
nltk.download('punkt_tab')

from nltk.tokenize import word_tokenize

# Tokenize the cleaned text
df['tokens'] = df['cleaned_text'].apply(word_tokenize)

# Display the first few rows of tokenized tweets
df.head()
````

# étape 3 Classification des Tweets . K-Means
Dans cette étape, nous allons appliquer l’algorithme K-Means pour regrouper les tweets similaires. Chaque groupe de tweets représentera un cluster, et l’utilisateur pourra lire un tweet représentatif de chaque groupe.

````python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

# Vectorize the cleaned text using TF-IDF
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(df['cleaned_text'])

# Nombre de clusters (k)
k = 5
kmeans = KMeans(n_clusters=k, random_state=42)
df['cluster'] = kmeans.fit_predict(X)

# Afficher les premières lignes avec les labels de clusters
df.head()
````

` K-Means pour classer les Tweets en k classes`

` Fuzzy C-Means pour le clustering flou :`on va le tester a la fin



````python
from sklearn.metrics import pairwise_distances

# Convertir les tweets en un ensemble de mots
def jaccard_distance(set1, set2):
    # Calculer l'intersection et l'union de deux ensembles
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))
    return 1 - (intersection / union)

# Exemple de calcul de la distance de Jaccard entre deux tweets (transformés en ensembles de mots)
tweet1 = set(df['tokens'][0])
tweet2 = set(df['tokens'][1])

# Calcul de la distance de Jaccard
distance = jaccard_distance(tweet1, tweet2)
print(f"Distance de Jaccard entre le tweet 1 et le tweet 2: {distance}")
````

# Choisir un Tweet par classe comme représentant
Une fois les clusters créés, vous devez sélectionner un tweet représentatif (centroïde) de chaque classe. Cela se fait généralement en sélectionnant le tweet le plus proche du centroïde du cluster (celui qui est le plus proche du centre de la classe)

````python
import numpy as np
from sklearn.metrics import pairwise_distances

# Fonction pour choisir un tweet représentatif pour chaque cluster
def get_representative_tweet(cluster_labels, cluster_id, tweets, vectorized_data):
    cluster_indices = np.where(cluster_labels == cluster_id)[0]
    cluster_vectors = vectorized_data[cluster_indices]

    # Calculer la distance entre chaque tweet et le centroïde du cluster
    centroid = np.mean(cluster_vectors, axis=0)
    distances = pairwise_distances(cluster_vectors, centroid.reshape(1, -1), metric='cosine')

    # Trouver l'indice du tweet le plus proche du centroïde
    closest_tweet_idx = np.argmin(distances)

    return tweets[cluster_indices[closest_tweet_idx]]  # Retourner le tweet le plus proche du centroïde

# Récupérer un tweet représentatif pour chaque cluster
representative_tweets = [get_representative_tweet(df['cluster'], i, df['Text'], X.toarray()) for i in range(k)]

# Afficher les tweets représentatifs pour chaque cluster
print("Tweets représentatifs par cluster:")
for i, tweet in enumerate(representative_tweets):
    print(f"Cluster {i}: {tweet}")
````

**Cluster 1 :**
Les tweets dans ce cluster semblent discuter de progrès en éducation, d'opportunités et d'événements récents. Cela inclut des sujets sur l'éducation, les développements professionnels, et les événements d'actualité ou de voyage.

 **Cluster 2 :**
Ce cluster contient des tweets plus orientés vers des sujets sociaux, politiques, et médiatiques. Par exemple, des tweets mentionnant des événements mondiaux, des discussions sur la politique, et des sujets d’actualité comme la santé publique, la politique, et les opinions publiques.

**Cluster 3 :**
Les tweets de ce cluster semblent plus centrés sur l’individu, la réflexion personnelle, et les défis de la vie. Les tweets mentionnent des expériences personnelles, des décisions de vie, et des aspirations futures.

# Visualisation des Clusters
Dans cette étape, nous allons réduire les dimensions des données pour les visualiser sur un graphique en 2D en utilisant PCA (Principal Component Analysis

````python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Reduce to 2 dimensions using PCA for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X.toarray())

# Plot the clusters
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=df['cluster'], cmap='viridis')
plt.title('Clustering of Tweets')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.colorbar(label='Cluster')
plt.show()
````

# Interprétation des Clusters
Dans cette section, nous allons examiner les termes clés de chaque cluster. Cela nous aidera à comprendre de quoi parlent les tweets dans chaque groupe.

````python
# Get the top terms for each cluster
terms = vectorizer.get_feature_names_out()

for i in range(k):
    print(f"Cluster {i}:")
    # Get the indices of the terms in the cluster's centroid (importance of terms)
    cluster_center = kmeans.cluster_centers_[i]
    sorted_terms_idx = cluster_center.argsort()[-10:][::-1]  # Top 10 terms
    top_terms = [terms[idx] for idx in sorted_terms_idx]
    print(top_terms)
    print()
````

````python
from sklearn.metrics.pairwise import cosine_similarity

# Calculer la similarité cosinus entre les tweets du même cluster
for i in range(k):
    # Extraire les tweets appartenant au cluster i
    cluster_tweets = X[df['cluster'] == i]

    # Calculer la matrice de similarité cosinus
    similarity_matrix = cosine_similarity(cluster_tweets)

    # Afficher la similarité entre les 5 premiers tweets du cluster
    print(f"Similarité pour les premiers tweets du cluster {i}:")
    print(similarity_matrix[:5, :5])  # Afficher les valeurs de similarité entre les 5 premiers tweets
    print()
````

# étape 3 Classification des Tweets Fuzzy C Means


 **Contrairement à K-Means, où chaque point appartient strictement à un cluster, Fuzzy C-Means permet à chaque point d’appartenir à plusieurs clusters avec un degré d’appartenance (entre 0 et 1).**

 *Calcul de la Distance de Jaccard entre les Tweets*

 La distance de Jaccard est utilisée pour mesurer la similarité entre deux ensembles. Dans le cas des tweets, les ensembles sont les mots dans chaque tweet.

 Cela peut être utilisé pour vérifier la similarité entre les tweets dans un même cluster.

````python
from fcmeans import FCM

# Appliquer Fuzzy C-Means (Fuzzy C Means)
fcm = FCM(n_clusters=k)
fcm.fit(X.toarray())  # Utilisez X.toarray() car FCM prend un tableau numpy
fcm_labels = fcm.predict(X.toarray())

# Ajouter les labels Fuzzy C-Means dans la DataFrame
# The fcm_labels already contain cluster assignments, so no need for argmax
df['fcm_cluster'] = fcm_labels
#df['fcm_cluster'] = fcm_labels.argmax(axis=0) # if fcm_labels was 2d probabilities

# Afficher les premières lignes avec les labels de clusters FCM
df.head()
````
